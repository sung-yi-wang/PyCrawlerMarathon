## Matplotlib 及繪圖流程介紹
- 認識 Matplotlib
- 瞭解 Matplotlib 繪圖流程

## 作業說明
- 以簡單範例資料繪圖，練習 Matplotlib 的畫板設置
  - 圖表標題 (title)
  - 圖表圖例 (legend)
  - 座標軸範圍 (axis)
  - 座標軸標題 (label)
  - 座標軸刻度字型、大小等 (ticks)
  - 圖表網格線 (grid)
  - 資料點標籤 (annotation)
  - 在圖表特定位置加上文字或箭頭 (text/arrow)
  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 可用的繪圖風格
print(plt.style.available)

# 設置繪圖風格
plt.style.use("fivethirtyeight")
```

```python
np.random.seed(888)

# 生成隨機資料點
x = np.random.randn(100)
y = np.random.randn(100)
# 繪製散點圖
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
```

### `title` 設置圖表標題
- `plt.title()`

```python
# 設定畫布大小
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
# 設置標題
plt.title('Random Scatter Plot')
```

### `legend` 設置圖表圖例
- `plt.legend()`

```python
# 為了讓畫布上有不同顏色的資料點，再生成一份相同數量的 y 軸資料點
# 生成隨機資料點
y1 = y.copy()
y2 = np.random.randn(100)

plt.figure(figsize=(8, 6))
# 繪製散點圖
plt.scatter(x, y1, label="True")  # label 是圖例上會出現的名稱
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
# 設置圖例
plt.legend(loc="upper right")
```

### 設置座標軸範圍
- `plt.axis()`
- `plt.xlim()` / `plt.ylim()`


```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
# 將座標移動為 x: 0-5, y: 0-10 的範圍
plt.axis([0, 5, 0, 10])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
# 將座標移動為 x: 0-5, y: 0-10 的範圍
plt.xlim([0, 5])
plt.ylim([0, 10])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
# 關閉座標軸
plt.axis("off") 
```

### `label` 設置座標軸標題
- `plt.xlabel()` / `plt.ylabel()`


```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
# 設置座標軸標題
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
```

### `ticks` 設置座標軸刻度字型、大小
- `plt.xticks()` / `plt.yticks()`

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
# 設置座標刻度樣式
plt.xticks(color="green", fontsize=12, rotation=45) # rotation 是旋轉角度
plt.yticks(color="blue", fontsize=12, rotation=75)
```

### `grid` 是否顯示圖表網格線
(某些繪圖風格已經有網格線，可以透過這個設置選擇是否顯示網格線）
- `plt.grid()`

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.xticks(color="green", fontsize=12, rotation=45)
plt.yticks(color="blue", fontsize=12, rotation=75)
# 關閉圖表網格線
plt.grid()
```

### `annotate` 設置資料點註釋
- `plt.annotate()`

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
plt.xlabel("This is X axis")
plt.ylabel("this is Y axis")
plt.xticks(color="green", fontsize=12, rotation=45)
plt.yticks(color="blue", fontsize=12, rotation=75)
# 設置資料點註釋
# Hint: xy 是註釋要出現的座標，xytext 是文字座標，arrowprops 是箭頭的樣式設定參數，facecolor 是箭頭的顏色
plt.annotate('Example Arrow', xy =(-1, -2), xytext =(-1, -2.8),  
            arrowprops = dict(facecolor ='yellow', shrink=0.05))
```

### `text` / `arrow` 設置圖表說明文字&箭頭
- `plt.text()` / `plt.arrow()`

```python
plt.figure(figsize=(8, 8))
plt.scatter(x, y1, label="True")
plt.scatter(x, y2, label="False")
plt.title("Random Scatter Plot")
plt.legend(loc="upper right")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.xticks(color="green", fontsize=12, rotation=45)
plt.yticks(color="blue", fontsize=12, rotation=75)
plt.annotate('Example Arrow', xy =(-1, -2), xytext =(-1, -2.8),  
            arrowprops = dict(facecolor ='yellow', shrink=0.05))
# 設置圖表說明文字
plt.text(1, -1, "Some text here", ha="left", family='serif', style='italic', size=12, rotation=10, wrap=True)
# 設置箭頭
plt.arrow(1, -1, 1, 1,  # 起始x座標, 起始y座標, 往x座標方向的長度, 往y座標方向的長度
          head_width=0.1, width=0.01, ec="green")  # 箭頭寬度, 箭身寬度, 顏色
```
